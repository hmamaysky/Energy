```python
import pandas as pd
```

```python
m = pd.read_csv("/shared/share_mamaysky-glasserman/energy_drivers/2023/DataProcessing/cosine/cosine.csv", 
                 header=0, index_col=0)
df = pd.read_excel('2018-05-04 energy word grouping 387 words.xlsx', index_col=0)
# Group by 'Topic' and aggregate the index (words) into lists
topic_to_words = df.groupby('Topic').apply(lambda x: list(x.index)).to_dict()

new_tokens = list(set(m.index).difference(df.index))
m_387 = m.loc[df.index, new_tokens]
assert m_387.shape == (387, 54)

df_avg_cs, df_diff_cs = pd.DataFrame(), pd.DataFrame()
# Difference between average cosine similarity with the existing tokens in topic i 
# and average across all seven topics
df_avg_cs['all'] = m_387.mean()
for topic_name, topic_tokens in topic_to_words.items():
    df_avg_cs[topic_name] = m_387.loc[topic_tokens].mean()
    df_diff_cs[topic_name] = df_avg_cs[topic_name] - df_avg_cs['all']
    
df_diff_cs
```

#### largest difference

```python
max_value_column = df_diff_cs.max().idxmax()
max_value_index = df_diff_cs[max_value_column].idxmax()
location_of_largest = (max_value_index, max_value_column)
location_of_largest
```

```python
print(f"For example, the candidate token {location_of_largest[0]}, \
which was not part of our original 387-token list, \
had an average cosine similarity with the existing tokens in topic {location_of_largest[1]} \
of {df_avg_cs.loc[location_of_largest]:.4f}, \
whereas its average co-occurrence across all seven topics was {df_avg_cs.loc[location_of_largest[0], 'all']:.4f}. \
The resultant difference of {df_diff_cs.loc[location_of_largest]:.4f} \
was the highest of all our candidate tokens."
     )
```

#### second largest difference

```python
sorted_values = df_diff_cs.unstack().sort_values(ascending=False).unique()
# Get the second largest value
second_largest_value = sorted_values[1]
location_of_second_largest = df_diff_cs[df_diff_cs == second_largest_value].stack().index[0]
location_of_second_largest
```

```python
print(f"For example, the candidate token {location_of_second_largest[0]}, \
which was not part of our original 387-token list, \
had an average cosine similarity with the existing tokens in topic {location_of_second_largest[1]} \
of {df_avg_cs.loc[location_of_second_largest]:.4f}, \
whereas its average co-occurrence across all seven topics was {df_avg_cs.loc[location_of_second_largest[0], 'all']:.4f}. \
The resultant difference of {df_diff_cs.loc[location_of_second_largest]:.4f} \
was the second highest of all our candidate tokens."
     )
```
